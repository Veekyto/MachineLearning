## 填充

| 参数              | 含义&输入                                                                                                                                                                  |
|-----------------|------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| missing_values  | 告诉SimpleImputer，数据中缺失值是什么，默认空值np.nan                                                                                                              |
| strategy        | <ul><li>我们填补缺失值的策略，默认均值。</li><li>输入`"mean"`使用均值填补（仅对数值特征的值）。</li><li>输入`"median"`使用中位值填补（仅对数值特征的值）。</li><li>输入`"most_frequent"`使用众数填补（对数值型和字符特征的值）。</li><li>输入`"constant"`表示常量，常量值需要选择`fill_value`中的值（对数值型和字符特征的值）。</li></ul> |
| fill_value      | 当参数strategy为`"constant"`时使用，输入法字符或数值来替换丢失的值，常用0                                                                                                 |
| copy            | 默认值为True，保持数据集去拷贝值的填补，反之减少内存消耗同样的数据。                                                                                                   |

```python
data=pd.read_csv(r"day08_Narrativedata.csv",index_col=0)
data.info()
```

```python
from sklearn.impute import SimpleImputer
imp_mean=SimpleImputer()
imp_median=SimpleImputer(strategy="median")
age=data.loc[:,"Age"].values.reshape(-1,1)
age
```

```python
imp_mean = imp_mean.fit_transform(age)
imp_median=imp_median.fit_transform(age)
imp_mean
```

```python
data.loc[:,"Age"]=imp_median
data.info()
```

```python
Embarked = data.loc[:,"Embarked"].values.reshape(-1,1)
imp_mode = SimpleImputer(strategy = "most_frequent")
data.loc[:,"Embarked"] = imp_mode.fit_transform(Embarked)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
y=data.iloc[:,-1]
y
```

```python
le=LabelEncoder()
le=le.fit(y)
le.classes_ 
label=le.transform(y)
label
```

```python
data.iloc[:,-1]=label
data.head()
```

```python
from sklearn.preprocessing  import OrdinalEncoder
OrdinalEncoder().fit(data.iloc[:,1:-1]).categories_
```

```python
from sklearn.preprocessing import OneHotEncoder
X = data.iloc[:,1:-1]
enc = OneHotEncoder(categories='auto').fit(X)
result = enc.transform(X).toarray()
result
```

```python
data=pd.concat([data,pd.DataFrame(result)],axis=1)
data.head
```

```python
data.drop(["Sex","Embarked"],axis=1,inplace=True)
```

# 特征选择

## 过滤法

方差

```python
from sklearn.feature_selection import VarianceThreshold
data=pd.read_csv(r"day08_digit recognizor.csv")
x=data.iloc[:,1:]
y=data.iloc[:,0]
x.info()
```

```python

selector=VarianceThreshold()

x_var0=selector.fit_transform(x)
x_var0.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
rfc=RandomForestClassifier(n_estimators=15,random_state=1)
cross_val_score(rfc,x,y,cv=10).mean()
```

```python
cross_val_score(rfc,x_var0,y,cv=11).mean()
```

### 卡方过滤

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
x_chi=SelectKBest(chi2,k=300).fit_transform(x,y)
x_chi.shape
```

```python
cross_val_score(rfc,x_chi,y,cv=11).mean()
```

### **F检验**  
它即可以做回归也
可以做分类，因此包含feature_selection.f_classif（F检验分类）和feature_selection.f_regression（F检验回
归）两个类。

```python
from sklearn.feature_selection import f_classif
F,p=f_classif(x,y)
k=F.shape[0]-(p>0.05).sum()
k
```

```python
x_f=SelectKBest(f_classif,k=k).fit_transform(x,y)
x_f.shape
```

### **互信息法**  
互信息法不返回p值或F值类似的统计量，它返回“每个特征与目标之间的互信息量的估计”，这个估计量在[0,1]之间
取值，为0则表示两个变量独立，为1则表示两个变量完全相关。

```python
from sklearn.feature_selection import mutual_info_classif
result = mutual_info_classif(x,y)
k=result.shape[0]-sum(result<=0)
k
```

### 过滤法总结

| 类 | 说明 | 超参数的选择 |
|---|---|---|
| VarianceThreshold | 方差过滤，可输入方差阈值，返回方差大于阈值的新特征矩阵 | 看具体数据究竟是含有更多噪声还是更多有效特征一般就使用0或1来筂选也可以画学习曲线或取中位数跑模型来帮助确认 |
| SelectKBest | 用来选取K个统计量结果最佳的特征，生成符合统计量要求的新特征矩阵 | 看配合使用的统计量 |
| chi2 | 卡方检验，专用于分类算法，捕捉相关性 | 追求p小于显著性水平的特征 |
| f_classif | F检验分类，只能捕捉线性相关性要求数据服从正态分布 | 追求p小于显著性水平的特征 |
| f_regression | F检验回归，只能捕捉线性相关性要求数据服从正态分布 | 追求 $p$ 小于显著性水平的特征 |
| mutual_info_classif | 互信息分类，可以捕捉任何相关性不能用于稀疏矩阵 | 追求互信息估计大于0的特征 |
| mutual_info_regression | 互信息回归，可以捕捉任何相关性不能用于稀疏矩阵 | 追求互信息估计大于0的特征 |

## 嵌入法

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier as RFC
RFC_ = RFC(n_estimators =10,random_state=0)
x_embedded = SelectFromModel(RFC_,threshold=0.005).fit_transform(x,y)
x_embedded.shape
```

## 包装法

```python
from sklearn.feature_selection import RFE
RFC_ = RFC(n_estimators =10,random_state=0)
selector = RFE(RFC_, n_features_to_select=340, step=50).fit(x, y)
selector.support_.sum()
```

```python
X_wrapper = selector.transform(x)
cross_val_score(RFC_,X_wrapper,y,cv=5).mean()
```
